```python
import pandas as pd
data = pd.read_csv('Levels_Fyi_Salary_Data.csv')
data
```

```python
data.info()
```

```python
# feature => Company, Age, Salary
# Label => Purchased
# Category => Company

# Best Practices
# 1. if label is null delete the record
# 2. if the feature is null?
 # a. speak to domain expert, take his recoomendation
 # b. if it is numeric => you can replace with mean
 # C. if it is alphanumeric and categorical => you can replace with mode
 # d. delete the record

data[data['Purchased'].isnull()]
```

```python
data.drop(data.index[data['Purchased'].isnull()], inplace=True)
data
```

```python
data[data['Salary'].isnull()]
```

```python
data['Salary'].fillna(data['Salary'].mean(), inplace=True)
data[data['Salary'].isnull()]
```
